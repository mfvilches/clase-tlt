```python
#leer archivo excel
olympics_data = pd.read_excel('../data/raw/olympics-data.xlsx', sheet_name='results')
#ejemplo de guardado excel a csv
olympics_data.to_csv('../data/raw/olympics-data.csv')
```

```python
olympics_data
```

```python
#Filtro en type que sea Summer
ol_data_summer = olympics_data[olympics_data['type']=='Summer']
ol_data_summer
```

```python
noc_regions = pd.read_csv('../data/raw/noc_regions.csv')
noc_regions
```

```python
# hacer un BUSCARV

results_new = pd.merge(ol_data_summer,noc_regions, left_on='noc',right_on='NOC',how='left')
results_new
```

```python
#filtros de columnas

lista_columnas = ['year','discipline','region']

results_filtrado = results_new[lista_columnas]
results_filtrado
```

```python
#filtrar columnas del primer archivo y agregar una del segundo archivo
lista_columnas_izquierdas = list(ol_data_summer.columns)
lista_columnas_izquierdas.append('region')
lista_columnas_izquierdas
```

```python
#ver el resultado filtrado
results_filtrado = results_new[lista_columnas_izquierdas]
results_filtrado
```

```python
#Guardar resultados en archivo Excel
results_filtrado.to_excel('../data/raw/resultados_filtrado.xlsx')
```
